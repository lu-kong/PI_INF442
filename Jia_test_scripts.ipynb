```python
df_2007 = pd.read_csv('E:/PI_INF442/p1-power-weather/cleaned_data/power_2007.csv', sep = ';')
df_2008 = pd.read_csv('E:/PI_INF442/p1-power-weather/cleaned_data/power_2008.csv', sep = ';')
df_2009 = pd.read_csv('E:/PI_INF442/p1-power-weather/cleaned_data/power_2009.csv', sep = ';')
df_2010 = pd.read_csv('E:/PI_INF442/p1-power-weather/cleaned_data/power_2010.csv', sep = ';')
```

```python
df_2007.head()
```

```python
df_2007 = time_rescale(df_2007, '1H')
df_2008 = time_rescale(df_2008, '1H')
df_2009 = time_rescale(df_2009, '1H')
df_2010 = time_rescale(df_2010, '1H')
```

```python
df_2007['full_time'].dt.hour
```

```python
d_2007 = df_2007.groupby(df_2007['full_time'].dt.hour).sum()
d_2008 = df_2008.groupby(df_2008['full_time'].dt.hour).sum()
d_2009 = df_2009.groupby(df_2009['full_time'].dt.hour).sum()
d_2010 = df_2010.groupby(df_2010['full_time'].dt.hour).sum()
```

```python
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (10,5))
x = d_2007.index
ax.plot(x, d_2007['g_active_power'], 'bo--')
ax.plot(x, d_2008['g_active_power'], 'ro--')
ax.plot(x, d_2009['g_active_power'], 'go--')
ax.plot(x, d_2010['g_active_power'], 'yo--')
ax.legend(['2007', '2008', '2009', '2010'])
ax.set_xlabel('hour of a day')
ax.set_ylabel('global active power')
plt.title('global active power hourly distribution in a day (during year 2007-2010)')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
x = d_2007.index
ax.plot(x, d_2007['g_reactive_power'], 'bo--')
ax.plot(x, d_2008['g_reactive_power'], 'ro--')
ax.plot(x, d_2009['g_reactive_power'], 'go--')
ax.plot(x, d_2007['g_reactive_power'], 'yo--')
ax.legend(['2007', '2008', '2009', '2010'])
ax.set_xlabel('hour of a day')
ax.set_ylabel('global active power')
plt.title('global active power hourly distribution in a day (during year 2007-2010)')
plt.show()
```

```python
def test(str = 'g_active_power'):
    fig, ax = plt.subplots(figsize = (10,5))
    x = d_2007.index
    ax.plot(x, d_2007[str], 'bo--')
    ax.plot(x, d_2008[str], 'ro--')
    ax.plot(x, d_2009[str], 'go--')
    ax.plot(x, d_2010[str], 'yo--')
    ax.legend(['2007', '2008', '2009', '2010'])
    ax.set_xlabel('hour in a day')
    ax.set_ylabel(str)
    plt.title(str+'  hourly distribution in a day (during year 2007-2010)')
    plt.show()
```
